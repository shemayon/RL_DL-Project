```python
import pickle

# Path to the CIFAR-100 train file
train_file = r"C:\Users\HUMAID\Downloads\cifar-100-python\cifar-100-python\train"

# Open the train file
with open(train_file, 'rb') as f:
    train_data = pickle.load(f, encoding='bytes')

# Access the data and labels
data = train_data[b'data']
labels = train_data[b'fine_labels']

# Print the data shape and label count
print("Data shape:", data.shape)
print("Number of labels:", len(labels))
```

```python
import pickle

# Path to the CIFAR-100 meta file
meta_file = r"C:\Users\HUMAID\Downloads\cifar-100-python\cifar-100-python\meta"

# Open the meta file
with open(meta_file, 'rb') as f:
    meta_data = pickle.load(f, encoding='bytes')

# Access the fine label names
fine_label_names = meta_data[b'fine_label_names']

# Print class labels and corresponding names
for i, name in enumerate(fine_label_names):
    print(f"Class Label {i}: {name.decode('utf-8')}")

# Access the coarse label names
coarse_label_names = meta_data[b'coarse_label_names']

# Print coarse class labels and corresponding names
for i, name in enumerate(coarse_label_names):
    print(f"Coarse Class Label {i}: {name.decode('utf-8')}")
```

```python
import random
import pickle

# Load the CIFAR-100 train dataset using pickle
with open('./dataset/cifar-100-python/train', 'rb') as file:
    train_data = pickle.load(file, encoding='bytes')

# Get a random index
random_index = random.randint(0, len(train_data[b'fine_labels']) - 1)

# Get the class label and image filename of the random example
class_label = train_data[b'fine_labels'][random_index]
image_filename = train_data[b'filenames'][random_index]

# Print the class label and image filename
print(f"Class Label: {class_label}")
print(f"Image Filename: {image_filename.decode('utf-8')}")
```
